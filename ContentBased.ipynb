```python
movie_df_merge = df_2.merge(df_1, on='movie_id')
movie_df_merge.head()
```

```python
movies_df_cleaned = movie_df_merge.drop(columns=['homepage','title_y','status','production_countries'])
```

```python
movie_df = movies_df_cleaned.rename(index=str,columns={"title_x":"title"})
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
 

tfv = TfidfVectorizer(min_df=3, max_features=None, strip_accents='unicode',analyzer='word',token_pattern=r'\w{1,}', ngram_range=(1,3), stop_words='english')
movie_df['overview'] = movie_df['overview'].fillna('')
```

```python
tfv_matrix = tfv.fit_transform(movie_df['overview'])
tfv_matrix
```

```python
tfv_matrix.shape
```

```python
from sklearn.metrics.pairwise import sigmoid_kernel

sig = sigmoid_kernel(tfv_matrix,tfv_matrix)
```

```python
sig[0]
```

```python
indices = pd.Series(movie_df.index, index=movie_df['original_title']).drop_duplicates()
```

```python
indices
```

```python
def get_recommendation(title, sig=sig):
    idx = indices[title]
    sig_scores = list(enumerate(sig[idx]))
    sig_scores = sorted(sig_scores, key=lambda x: x[1], reverse=True)
    sig_scores= sig_scores[1:11]
    movie_indices = [i[0] for i in sig_scores]
    return movie_df['original_title'].iloc[movie_indices]
```
